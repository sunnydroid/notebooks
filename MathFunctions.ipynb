# Log function

Ranges from (greater than 0) to  $ \infty $

`log` of 0 and negative numbers is undefined or NaN

**Input is limited to positive numbers and ouput plateaus for larger numbers**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.1, 1000, 0.1)
plt.figure(figsize=(10, 6))
plt.plot(x, np.log(x))
plt.xlabel('x')
plt.ylabel('log(x)')
plt.show()
```

# Exponent function

Ranges from $ -\infty $ to $+\infty $

**Output is limited to 0 for negative numbers and grows exponentially for postive numbers**

```python
x = np.arange(-10, 100, 0.1)
plt.figure(figsize=(10, 6))
plt.plot(x, np.exp(x))
plt.xlabel('x')
plt.ylabel('exp(x)')
plt.show()
```
