```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create rand normal matrix
np_matrix = np.random.randn(1000, 4)
np_matrix
```

```python
print('mean of 1st col {}'.format(np_matrix[:, 0].mean()))
print('sd of 1st col {}'.format(np_matrix[:, 0].std()))
print('mean of 2nd col {}'.format(np_matrix[:, 1].mean()))
print('sd of 2nd col {}'.format(np_matrix[:, 1].std()))
print('mean of 3rd col {}'.format(np_matrix[:, 2].mean()))
print('sd of 3rd col {}'.format(np_matrix[:, 2].std()))
print('mean of 4th col {}'.format(np_matrix[:, 3].mean()))
print('sd of 4th col {}'.format(np_matrix[:, 3].std()))
```

```python
pd_df = pd.DataFrame(np_matrix, columns=['A', 'B', 'C', 'D'])
pd_df.head()
```

```python
# plot scatter matrix of the dataset
pd.plotting.scatter_matrix(pd_df, alpha=0.3, figsize=(9,9))
```
